```python
df_dicts = pd.DataFrame() 
for i in range(len(data)):

    remove_str = data[i].find("<Response [200]>,")
    new_data = data[i][remove_str:].replace('<Response [200]>,', "").replace('""', '"').replace('"{', "{").replace("\n", "").strip().replace('}"', '}')

    new_data = json.loads(new_data)["items"]
    
    df_dicts = df_dicts.append(new_data)
    
     if i % 1000 == 0:
        print("Linhas Processadas", i)
    
```

```python
df_dicts.shape
```

```python
df_dicts.id.unique().shape
```

```python
df_dicts.head(10)
```

```python
df_dicts["inputDateTime"] = pd.to_datetime(df_dicts.inputDateTime)
df_dicts["outputDateTime"] = pd.to_datetime(df_dicts.outputDateTime)
```

```python
def convert_dias_semana(dia_semana):
    case = {
       'Monday': 'Segunda-Feira', 
       'Sunday': 'Domingo', 
       'Saturday': 'Sabado', 
       'Friday': 'Sexta-Feira', 
       'Wednesday': 'Quarta-Feira', 
       'Tuesday': 'Terça-Feira',
       'Thursday': 'Quinta-Feira'
    }
    
    return case.get(dia_semana)


def convert_dias_to_int(dia_semana):
    case = {
        'Domingo': 1, 
        'Segunda-Feira':2 , 
        'Terça-Feira': 3,
        'Quarta-Feira': 4, 
        'Quinta-Feira': 5,
        'Sexta-Feira': 6, 
        'Sabado': 7
    }
    
    return case.get(dia_semana)



df_dicts["DIA_DA_SEMANA"] = df_dicts.inputDateTime.dt.day_name().apply(lambda x: convert_dias_semana(x))
df_dicts["HORA_TRANSACAO"] = df_dicts.inputDateTime.dt.hour
df_dicts["INDEX_DIA_DA_SEMANA"] = df_dicts["DIA_DA_SEMANA"].apply(lambda x: convert_dias_to_int(x))
df_dicts["DIA_DO_MES"] = df_dicts.inputDateTime.dt.day
df_dicts["MES"] = df_dicts.inputDateTime.dt.month
df_dicts["ANO"] = df_dicts.inputDateTime.dt.year
```

```python
df_dicts.head(10)
```

```python
string_find = "^Facebook"
df_facebook = df_dicts[df_dicts["establishmentName"].str.contains(string_find)]
```

### CRIA E POPULA COLUNA RETENTATIVAS E TENTOU MAIS DE UMA VEZ

OBS: Considerando que a cobrança é mensal

    1: Para Retentativas

    0: Para não Retentativas
    
    1: Para tentou mais de uma vez
    
    0: Para quem não tentou mais de uma vez
    
* Para a coluna RETENTATIVAS, o primeiro valor pode ser 0, pois é a primeira transação e os demais valores, serem iguais a 1, pois reprenta que houve varias RETENTATIVAS após a primeira. 

* A coluna TENTOU_MAIS_DE_UMA_VEZ = 0, significa que só houve uma tentavia de cobrança daquela transação. TENTOU_MAIS_DE_UMA_VEZ = 1, indica que houve mais de uma tentativa. 

```python
df_facebook["RETENTATIVAS"] = 1
df_facebook["TENTOU_MAIS_DE_UMA_VEZ"] = 1
```

```python
df_facebook.idCard.value_counts()
```

```python
select_columns = ["idCard", "inputDateTime", "MES", "ANO", ]
groupby_columns = ["idCard",  "MES", "ANO", ]
df_facebook_distincted = df_facebook[select_columns].groupby(groupby_columns).min().reset_index()
```

```python
condicao_1 = (df_facebook_distincted.idCard == 255928959) & \
                        (df_facebook_distincted.MES == 11) & \
                        (df_facebook_distincted.ANO == 2020)


df_facebook_distincted[condicao_1]
```

```python
for i, row in df_facebook_distincted.iterrows():
    
    idcard = row["idCard"]
    mes = row["MES"]
    ano = row["ANO"]
    date_time = row["inputDateTime"]
    
    
    condicoes_1 = (
                        (df_facebook.idCard == idcard) & \
                        (df_facebook.MES == mes) & \
                        (df_facebook.ANO == ano) & \
                        (df_facebook.inputDateTime == date_time)
    )
    
       
    df_facebook.loc[ condicoes_1 , "RETENTATIVAS"] = 0
    
    
    condicoes_2 = (
                        (df_facebook.idCard == idcard) & \
                        (df_facebook.MES == mes) & \
                        (df_facebook.ANO == ano) 
    ) 
    
    
    # É porque ele tentou uma vez e deu certo
    if df_facebook[condicoes_2].shape[0] == 1:
        df_facebook.loc[ condicoes_2 , "TENTOU_MAIS_DE_UMA_VEZ"] = 0
        
    if i % 5000 == 0:
        print(i)
    
```

### NÚMERO MAXIMO DE TENTATIVAS

```python
import plotly.express as px
```

```python
select_columns = ["terminalDescription", "RETENTATIVAS"]
groupby_columns = ["terminalDescription"]
df_max_tentativas_por_status = df_facebook[select_columns].groupby(groupby_columns).sum().reset_index().sort_values("RETENTATIVAS", ascending=False)
```

```python
fig = px.bar(df_max_tentativas_por_status, y='RETENTATIVAS', x='terminalDescription', text_auto='.2s',
            title="Quantidade maxima de tentativas agrupados por status do cartão")
fig.show()
```

```python
df_nao_tentaram = df_facebook[df_facebook.TENTOU_MAIS_DE_UMA_VEZ == 0][["terminalDescription", "TENTOU_MAIS_DE_UMA_VEZ"]].groupby(["terminalDescription", "TENTOU_MAIS_DE_UMA_VEZ"]).size().reset_index().rename(columns={0:"QT_NAO_TENTARAM_MAIS_DE_UMA_VEZ"}).drop("TENTOU_MAIS_DE_UMA_VEZ", axis=1).sort_values("QT_NAO_TENTARAM_MAIS_DE_UMA_VEZ", ascending=False)
```

```python
fig = px.bar(df_nao_tentaram, y='QT_NAO_TENTARAM_MAIS_DE_UMA_VEZ', x='terminalDescription', text_auto='.2s',
            title="Quantidade de transações que não tentaram mais de uma vez")
fig.show()
```

#### COMPARANDO OS STATUS DAS TENTATIVAS DE PAGAMENTOS DE 2020 X 2021

```python
select_columns = ["terminalDescription", "MES", "ANO", "RETENTATIVAS"]
groupby_columns = [ "terminalDescription", "MES", "ANO", ]


df_facebook[select_columns].groupby(groupby_columns).sum().reset_index().sort_values("RETENTATIVAS", ascending=False)
```

```python
df_stats = df_facebook[select_columns].groupby(groupby_columns).sum().reset_index()
df_stats["ANO_MES"] = df_stats.ANO.astype(str) + df_stats.MES.apply(lambda x: '-0'+str(x) if len(str(x)) == 1 else str(x))
```

```python
df_stats.sort_values(["ANO_MES", "RETENTATIVAS"])
```

```python
fig = px.bar(df_stats.sort_values(["ANO_MES", "RETENTATIVAS"]), x="ANO_MES", y="RETENTATIVAS", color="terminalDescription", 
            title="Número de retentativas 2020 X 2021 por STATUS do cartão")
fig.update_xaxes(tickangle=45)
fig.show()
```

```python
q=[0, .2, .4, .6, .8, 1]
cat, bins = pd.qcut(df_facebook.amount, q=q, retbins=True, precision=0)
```

```python
df_facebook["FAIXA_AMOUNT"] = cat.apply(lambda x: str(x).replace("(", "").replace("]", "").replace(".0", "").replace(",", " a").replace("-1", "0"))
```

```python
df_faixa_amount = df_facebook[["terminalDescription", "FAIXA_AMOUNT"]].groupby(["FAIXA_AMOUNT", "terminalDescription"]).size().reset_index().rename(columns={0:"FREQUENCIA"})
```

```python
df_faixa_amount.head(10)
```

```python
fig = px.bar(df_faixa_amount.sort_values(["FAIXA_AMOUNT", "FREQUENCIA"]), x="FAIXA_AMOUNT", y="FREQUENCIA", color="terminalDescription",
            title="FRÊQUENCIA DE TRANSAÇÕES NEGADAS POR MOTIVO E POR FAIXA DE CUSTO")
fig.update_xaxes(tickangle=45)
fig.show()
```

#### No Gráfico acima, podemos ver na ultima faixa, 1250 a 18714, que o STATUS "SEM LIMITE" tem uma frequência maior de aparecimento em relação a o STATUS "BLOQUEADO_CARTÃO" e um outro ponto importante, é que na cor rosa que representa o STATUS "LIMITE_DIARIO_CONTA_CONTROLE_ULTRAPASSADO", tem uma frequência maior na medida em que a FAIXA de AMOUNT aumenta.

```python
QT_TRANSACOES_NA_SEMANA = df_facebook[["id", "INDEX_DIA_DA_SEMANA", "DIA_DA_SEMANA"]].groupby(["INDEX_DIA_DA_SEMANA", "DIA_DA_SEMANA"]).count().reset_index().rename(columns={"id": "QT_TRANSACOES"}).sort_values("INDEX_DIA_DA_SEMANA", ascending=True)
```

```python
QT_TRANSACOES_NA_SEMANA
```

```python
fig = px.line(QT_TRANSACOES_NA_SEMANA, x="DIA_DA_SEMANA", y="QT_TRANSACOES", title='Número de transações ao longo da semana')
fig.show()
```

```python
QT_TRANSACOES_POR_HORA = df_facebook[["id", "HORA_TRANSACAO"]].groupby("HORA_TRANSACAO").count().reset_index().rename(columns={"id": "QT_TRANSACOES"}).sort_values("HORA_TRANSACAO")
```

```python
fig = px.line(QT_TRANSACOES_POR_HORA, x="HORA_TRANSACAO", y="QT_TRANSACOES", title='Número de transações ao longo do dia')
fig.show()
```

```python
q=[.0, .2, .4, .6, .8, 1]
cat, bins = pd.qcut(df_facebook.DIA_DO_MES, q=q, retbins=True, precision=0)
```

```python
df_facebook["FAIXA_DIA_DO_MES"] = cat.apply(lambda x: str(x).replace("(", "").replace("]", "").replace(".0", "").replace(",", " a").replace("0 a", "1 a"))
```

```python
df_faixa_dia_mes = df_facebook[["terminalDescription", "FAIXA_DIA_DO_MES"]].groupby(["FAIXA_DIA_DO_MES", "terminalDescription"]).size().reset_index().rename(columns={0:"FREQUENCIA"})
```

```python
df_faixa_dia_mes.head(10)
```

```python
fig = px.bar(df_faixa_dia_mes.sort_values(["FAIXA_DIA_DO_MES", "FREQUENCIA"]), x="FAIXA_DIA_DO_MES", y="FREQUENCIA", color="terminalDescription",
            title="QUANTIDADE TRANSAÇÕES NEGADAS POR FAIXA DE DIAS E POR MOTIVOS")
fig.update_xaxes(tickangle=45)
fig.show()
```

#### Com base no gráfico acima, para faixa 1 a 7 os indicadores BLOQUEADO_STATUS_CARTAO, SEM_LIMITE, LIMITE_DIARIO_CONTA_CONTROLE_ULTRAPASSADO tem um aumento considerável comparado as outras faixas de dias (7 a 13, 13 a 18 e 25 a 31), indicando que, no inicio do mês, as transações aumentam bastante e a partir do dia 7 de cada mês, há uma leve caída e volta a aumentar na faixa de 18 a 25. Pode ser por conta do vencimento da fatura, pois os vencimentos geralmente variam entre dia 5 a 8 ou dia 15 a 20 de cada mês e portanto, os níveis de cobrança, aumentam consideravelmente nesses periodos.
